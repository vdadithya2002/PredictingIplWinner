```python
def simulate_season(matches_df, base_points, n_simulations=10000):
    teams = base_points.keys()
    playoff_counts = defaultdict(int)
    win_counts = defaultdict(int)

    for _ in range(n_simulations):
        table = {team: {'Pts': base_points[team]['Pts'], 'Mat': base_points[team]['Mat']} for team in teams}

        for _, row in matches_df.iterrows():
            team1 = row['Team 1']
            team2 = row['Team 2']
            p_team1 = row['Team1_Win_Probability']

            winner = np.random.choice([team1, team2], p=[p_team1, 1 - p_team1])
            table[winner]['Pts'] += 2
            table[team1]['Mat'] += 1
            table[team2]['Mat'] += 1

        # Convert table to DataFrame
        standings = pd.DataFrame.from_dict(table, orient='index')
        standings['Team'] = standings.index
        standings = standings.sort_values(by=['Pts'], ascending=False).reset_index(drop=True)

        # Record playoff appearances and winner
        top_4 = standings.head(4)['Team'].tolist()
        for team in top_4:
            playoff_counts[team] += 1
        win_counts[standings.iloc[0]['Team']] += 1

    return playoff_counts, win_counts
```

```python
import pandas as pd
import numpy as np
from collections import defaultdict

def simulate_playoffs(top4, match_prob_dict):
    t1, t2, t3, t4 = top4

    # Qualifier 1: t1 vs t2
    q1_winner = np.random.choice([t1, t2], p=[match_prob_dict.get((t1, t2), 0.5), 1 - match_prob_dict.get((t1, t2), 0.5)])
    q1_loser = t2 if q1_winner == t1 else t1

    # Eliminator: t3 vs t4
    elim_winner = np.random.choice([t3, t4], p=[match_prob_dict.get((t3, t4), 0.5), 1 - match_prob_dict.get((t3, t4), 0.5)])

    # Qualifier 2: q1_loser vs elim_winner
    q2_winner = np.random.choice([q1_loser, elim_winner], p=[match_prob_dict.get((q1_loser, elim_winner), 0.5), 1 - match_prob_dict.get((q1_loser, elim_winner), 0.5)])

    # Final: q1_winner vs q2_winner
    final_winner = np.random.choice([q1_winner, q2_winner], p=[match_prob_dict.get((q1_winner, q2_winner), 0.5), 1 - match_prob_dict.get((q1_winner, q2_winner), 0.5)])

    return final_winner
```

```python
def simulate_full_season_with_playoffs(matches_df, base_points, match_prob_dict, n_simulations=10000):
    teams = base_points.keys()
    playoff_counts = defaultdict(int)
    win_counts = defaultdict(int)

    for _ in range(n_simulations):
        table = {team: {'Pts': base_points[team]['Pts'], 'Mat': base_points[team]['Mat']} for team in teams}

        for _, row in matches_df.iterrows():
            team1 = row['Team 1']
            team2 = row['Team 2']
            p_team1 = row['Team1_Win_Probability']

            winner = np.random.choice([team1, team2], p=[p_team1, 1 - p_team1])
            table[winner]['Pts'] += 2
            table[team1]['Mat'] += 1
            table[team2]['Mat'] += 1

        # Convert to DataFrame and sort
        standings = pd.DataFrame.from_dict(table, orient='index')
        standings['Team'] = standings.index
        standings = standings.sort_values(by=['Pts'], ascending=False).reset_index(drop=True)

        # Get top 4 and record playoff appearance
        top4 = standings.head(4)['Team'].tolist()
        for team in top4:
            playoff_counts[team] += 1

        # Simulate playoffs and record winner
        winner = simulate_playoffs(top4, match_prob_dict)
        win_counts[winner] += 1

    return playoff_counts, win_counts
```

```python
import random
import networkx as nx
import matplotlib.pyplot as plt

# Sample match_prob dictionary as calculated earlier
match_prob = {
    ('GT', 'RCB'): 0.55,  # GT vs RCB in Qualifier 1
    ('RCB', 'GT'): 0.45,  # RCB vs GT in Qualifier 1
    ('PBKS', 'MI'): 0.52, # PBKS vs MI in Eliminator
    ('MI', 'PBKS'): 0.48, # MI vs PBKS in Eliminator
    ('GT', 'PBKS'): 0.56, # GT vs PBKS in Qualifier 2
    ('PBKS', 'GT'): 0.44, # PBKS vs GT in Qualifier 2
    ('GT', 'MI'): 0.60,   # GT vs MI in Qualifier 2
    ('MI', 'GT'): 0.40,   # MI vs GT in Qualifier 2
    ('RCB', 'PBKS'): 0.58,# RCB vs PBKS in Qualifier 2
    ('PBKS', 'RCB'): 0.42,# PBKS vs RCB in Qualifier 2
    ('RCB', 'MI'): 0.54,  # RCB vs MI in Qualifier 2
    ('MI', 'RCB'): 0.46   # MI vs RCB in Qualifier 2
}

# Simulate the match outcome based on probabilities
def simulate_match(team1, team2, prob_team1_win):
    winner = team1 if random.random() < prob_team1_win else team2
    return winner

# Function to simulate a playoff tree structure
def simulate_playoffs_tree(top4_teams, match_prob, num_simulations=10000):
    playoff_results = {'GT': 0, 'RCB': 0, 'PBKS': 0, 'MI': 0}  # Initialize win counts for each team

    for _ in range(num_simulations):
        # Simulate the first qualifier (GT vs RCB)
        winner_qualifier1 = simulate_match('GT', 'RCB', match_prob[('GT', 'RCB')])

        # Simulate the eliminator (PBKS vs MI)
        winner_eliminator = simulate_match('PBKS', 'MI', match_prob[('PBKS', 'MI')])

        # Simulate the second qualifier (Loser of Qualifier 1 vs Winner of Eliminator)
        loser_qualifier1 = 'RCB' if winner_qualifier1 == 'GT' else 'GT'
        winner_qualifier2 = simulate_match(loser_qualifier1, winner_eliminator, match_prob[(loser_qualifier1, winner_eliminator)])

        # Simulate the final (Winner of Qualifier 1 vs Winner of Qualifier 2)
        winner_final = simulate_match(winner_qualifier1, winner_qualifier2, match_prob[(winner_qualifier1, winner_qualifier2)])

        # Track the winner of the final
        playoff_results[winner_final] += 1

    return playoff_results

# Simulate playoffs
playoff_results = simulate_playoffs_tree(['GT', 'RCB', 'PBKS', 'MI'], match_prob)

# Calculate playoff probability and winning probability
total_simulations = 10000
playoff_probabilities = {team: playoff_results[team] / total_simulations for team in playoff_results}
```

```python
import random

# ... (Your previous code for match_prob, simulate_match, simulate_playoffs_tree) ...

def visualize_playoff_tree(top4_teams, match_prob, num_simulations=1000):
    """
    Simulates playoffs and prints a tree-like visualization.
    """

    for _ in range(num_simulations):
        print("-" * 30)  # Separator for each simulation

        # Qualifier 1
        winner_qualifier1 = simulate_match(top4_teams[0], top4_teams[1], match_prob[(top4_teams[0], top4_teams[1])])
        print(f"Qualifier 1: {top4_teams[0]} vs {top4_teams[1]} -> Winner: {winner_qualifier1}")

        # Eliminator
        winner_eliminator = simulate_match(top4_teams[2], top4_teams[3], match_prob[(top4_teams[2], top4_teams[3])])
        print(f"Eliminator: {top4_teams[2]} vs {top4_teams[3]} -> Winner: {winner_eliminator}")

        # Qualifier 2
        loser_qualifier1 = top4_teams[1] if winner_qualifier1 == top4_teams[0] else top4_teams[0]
        winner_qualifier2 = simulate_match(loser_qualifier1, winner_eliminator, match_prob[(loser_qualifier1, winner_eliminator)])
        print(f"Qualifier 2: {loser_qualifier1} vs {winner_eliminator} -> Winner: {winner_qualifier2}")

        # Final
        winner_final = simulate_match(winner_qualifier1, winner_qualifier2, match_prob[(winner_qualifier1, winner_qualifier2)])
        print(f"Final: {winner_qualifier1} vs {winner_qualifier2} -> **IPL Winner: {winner_final}**")

# Example usage
print("Visualizing the Playoff Simulation Output")
visualize_playoff_tree(['GT', 'RCB', 'PBKS', 'MI'], match_prob, num_simulations=5)
```
